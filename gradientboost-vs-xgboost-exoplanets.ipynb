```python
import pandas as pd
import numpy as np

data = pd.read_csv('../input/kepler-labelled-time-series-data/exoTrain.csv')
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

import time

start = time.time()

gbc = GradientBoostingClassifier(n_estimators=100, max_depth=2, random_state=2)
gbc.fit(X_train, y_train)
y_preds = gbc.predict(X_test)
score = accuracy_score(y_preds, y_test)
print('Score:', score)
end = time.time()
elapsed = end - start
print('Run time: ' + str(elapsed) + 'seconds')
```

```python
start = time.time()

xg_clf = XGBClassifier(n_estimators=100, max_depth=2, random_state=2)
xg_clf.fit(X_train, y_train)
y_preds = xg_clf.predict(X_test)
scores = accuracy_score(y_preds, y_test)
print('Score:', score)
end = time.time()
elapsed = end - start
print('Run time: '+str(elapsed) + 'seconds')
```
