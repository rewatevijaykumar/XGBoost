```python
import pandas as pd
import numpy as np
from sklearn import datasets
X,y = datasets.load_diabetes(return_X_y=True)    

```

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

xgb = XGBRegressor(booster='gbtree', objective='reg:squarederror', n_estimators=100, learning_rate=0.1, 
                     max_depth=6, random_state=2, n_jobs=-1)
scores = cross_val_score(xgb, X,y, cv=5, scoring='neg_mean_squared_error')
rmse = np.sqrt(-scores)
print('RMSE:',np.round(rmse,3))
print('RMSE mean: {:.3f}'.format(rmse.mean()))
```
